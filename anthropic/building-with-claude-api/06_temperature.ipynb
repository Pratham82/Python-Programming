```python
from helpers import add_user_message, add_assistant_message, chat

# Temperature controls randomness in model responses
# Lower values (e.g., 0.0) = more focused and deterministic
# Higher values (e.g., 1.0) = more creative and random

messages = []
add_user_message(messages, "Give me a short creative story about a cat.")

print("=== Temperature = 0.0 (Focused) ===")
answer = chat(messages, temperature=0.0)
print(answer)
print()

print("=== Temperature = 1.0 (Creative) ===")
messages2 = []
add_user_message(messages2, "Give me a short creative story about a cat.")
answer = chat(messages2, temperature=1.0)
print(answer)
```
